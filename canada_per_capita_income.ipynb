## Linear Regression

```python
import pandas as pd
import numpy as np
from sklearn import linear_model
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv("canada_per_capita_income.csv")
dataset
```

```python
len(dataset)
```

```python
dataset.head()
```

```python
dataset.shape
```

```python
plt.title("Scatter plot of Canada's per capita income (US$)")
plt.xlabel("year ")
plt.ylabel("per_capita_income")
plt.scatter(dataset["year"], dataset["per_capita_income"], alpha = 0.7)
plt.show()
```

```python
df = dataset.drop('per_capita_income' , axis = 'columns')
df
```

```python
income = dataset.per_capita_income
income
```

```python
# Create linear regression object
reg = linear_model.LinearRegression()
reg.fit(df, income)
```

```python
reg.predict([[2020]])
```

```python
reg.coef_
```

```python
reg.intercept_
```
